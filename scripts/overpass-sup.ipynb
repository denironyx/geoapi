```python
import geopandas as gpd
import pandas as pd

# Read the GeoDataFrame from the .osm file
df = gpd.read_file("../data/supermarket-export.osm", engine="pyogrio")

# Split the 'other_tags' column by comma to create separate entries for each key-value pair
tags_split = df['other_tags'].str.split(',')

# Create an empty list to store dictionaries of key-value pairs
tag_dicts = []

# Iterate over each entry in tags_split
for entry in tags_split:
    # Create an empty dictionary to store key-value pairs for the current entry
    tag_dict = {}
    # Iterate over each key-value pair in the entry
    for pair in entry:
        # Split the pair by '=>' to separate key and value
        key_value = pair.split('=>')
        # Check if the key-value pair has the expected format
        if len(key_value) == 2:
            # Remove surrounding quotes from the key and value
            key = key_value[0].strip('"')
            value = key_value[1].strip('"')
            # Add the key-value pair to the dictionary
            tag_dict[key] = value
    # Add the dictionary of key-value pairs to the list
    tag_dicts.append(tag_dict)

# Convert the list of dictionaries to a DataFrame
tags_df = pd.DataFrame(tag_dicts)

# Join the flattened tags DataFrame with the original GeoDataFrame
df = pd.concat([df, tags_df], axis=1)

# Drop the original 'other_tags' column
df.drop(columns=['other_tags'], inplace=True)

# Now, df contains each key from the 'other_tags' column as a separate column
```
